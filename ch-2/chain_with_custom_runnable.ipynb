Chain with custom runnable

```python
# Task 1

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a helpful assistant"),
        ("human","{input}")
    ])
```

```python
#Task 2

llm_openai = ChatOpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=os.environ.get("OPENROUTER_API_KEY"),
    model="nvidia/nemotron-nano-9b-v2:free",
    temperature=0
)
```

```python
#Task 3

str_parser = StrOutputParser() 
```

```python
# Task 4

from langchain_core.runnables import RunnableLambda

def dictionary_maker(text: str) -> dict:
    return {"text": text}

dictionary_maker_runnable = RunnableLambda(dictionary_maker)    
```

```python
# Task 5

prompt_post = ChatPromptTemplate.from_messages(
    messages = [
        ("system","you are a social media post generator "),
        ("human","Create a post for the text {text}")
    ]
)
```

```python
#task 6

llm_openai = ChatOpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=os.environ.get("OPENROUTER_API_KEY"),
    model="nvidia/nemotron-nano-9b-v2:free",
    temperature=0
)

```

```python
#Task 7

str_parser = StrOutputParser() 
```

```python
chain = prompt | llm_openai | str_parser | dictionary_maker | prompt_post | llm_openai | str_parser 

chain.invoke("what is the capital of france")
```
