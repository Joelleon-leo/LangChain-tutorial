Messages

```python
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage
import os
from langchain_openai import ChatOpenAI

llm_openai = ChatOpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=os.environ.get("OPENROUTER_API_KEY"),
    model="nvidia/nemotron-nano-9b-v2:free",
    temperature=0
)

my_messages = [
    SystemMessage(content="You are a gen-z assistant.who always answer in afun way."),
    HumanMessage(content="role of a student?")
    ]

llm_openai.invoke(my_messages).content
```

Prompts:

    these are messages which are sent to the llm

```python

from langchain_core.prompts import PromptTemplate

user_input = input("What do you want to ask?")

dynamic_prompt = PromptTemplate.from_template("write a fun fact about {topic}  in one sentence")

structured_prompt = dynamic_prompt.invoke({"topic": user_input})


llm_openai.invoke(structured_prompt).content
```

```python
from langchain_core.prompts import ChatPromptTemplate

Prompt_Template = ChatPromptTemplate.from_messages([
    ("system", "You are a {tone} assistant.who always answer in a {tone} way."),
    ("human", "write a fun fact about {topic}  in one sentence" )
])

user_tone = input("What tone do you want?")
user_input = input("What do you want to ask?")

read_prompt = Prompt_Template.invoke({"tone": user_tone, "topic": user_input})

llm_openai.invoke(read_prompt).content
```
