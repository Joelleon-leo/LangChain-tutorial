*Guiding in prompts*

```python
from langchain_openai import ChatOpenAI
import os

if os.getenv("OPENAI_API_KEY"):
    print("OPENAI_API_KEY environment variable is set")
    
```

```python
llm_openai = ChatOpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=os.environ.get("OPENROUTER_API_KEY"),
    model="nvidia/nemotron-nano-9b-v2:free",
    temperature=0
)
```

```python

from langchain_core.prompts import PromptTemplate

llm_openai.invoke("Tell me a joke.Generate the output in JSON format with a 'joke' key and the value being the joke ").content
```

Structured Output Using Pydantic Modals

```python
from pydantic import BaseModel,Field

class llm_schema(BaseModel):
    joke: str = Field(description="The joke generated by the model")
    Worth_Reading: bool = Field(description="Whether the joke is worth reading or not") 
```

```python
llm_structured_output = llm_openai.with_structured_output(llm_schema)

result=llm_structured_output.invoke("Tell me a joke.")
 
print(result)
```

Using TypedDict

```python
from typing import TypedDict
class llm_schema_dict(TypedDict):
    joke: str
    Worth_Reading: bool
```

```python
object = llm_schema_dict(NoJoke="Why did the scarecrow win an award? Because he was outstanding in his field!", Worth_Reading=True)
object
```
